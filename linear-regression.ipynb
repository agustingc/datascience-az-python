Simple Linear Regression


```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Import CSV to panda's DataFrame
data = pandas.read_csv('P12-SalaryData.csv')

# Create python list for x, y
x = np.array(data['YearsExperience'].values)
y = np.array(data['Salary'].values)

# create scatter plot
#plt.plot(x,y,'*')
plt.scatter(x,y)

# slope, intercept
model = np.polyfit(x,y,1)
m, b = model # destructure list

# add line of best fit
plt.plot(x, m*x + b)

#r-squared
corr_matrix = np.corrcoef(x,y)
corr_xy = corr_matrix[0,1]
r_squared = corr_xy**2
print(r_squared)

# validate r-squared calc with sklearn lib
predict = np.poly1d(model)
r2_score(y, predict(x)) #print


```
